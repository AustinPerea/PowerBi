## Violin plots
In this Lab activity, we will compare IQ scores among different test groups using violin plots.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mydata = pd.read_csv("scores.csv")
```

```python
# Create lists of IQ scores for group_a, group_b, group_c and group_d
group_a = mydata["group_a"].values.tolist()
group_b = mydata["group_b"].values.tolist()
group_c = mydata["group_c"].values.tolist()
group_d = mydata["group_d"].values.tolist()
```

```python
data = pd.DataFrame({'Groups': ['Group A'] * len(group_a) + ['Group B'] * len(group_b) + ['Group C'] * len(group_c) + ['Group D'] * len(group_d),
                     'IQ score': group_a + group_b + group_c + group_d})
```

<div class="alert alert-block alert-info" style="margin-top: 20px">
2. Create a violin plot for each of the IQ scores of different test groups using Seaborn's violinplot function. Use the whitegrid style and remove all axes splines, except the one on the bottom. Add a title.
</div>

```python
# Create figure
plt.figure(dpi=100)
sns.set (style = 'whitegrid')
ax = sns.violinplot(data=mydata)

ax.set_title("IQ scores for different test groups ")
ax.set_xlabel("Groups")
ax.set_ylabel("IQ Score")
sns.despine(top=True, right=True, left=True )
# Show plot
plt.show()
```
